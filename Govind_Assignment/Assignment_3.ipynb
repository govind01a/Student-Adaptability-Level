#### Handling Outliers

```python
quantile=df.quantile(q=[0.75,0.25],numeric_only=True)
quantile
```

```python
quantile.iloc[0]
```

```python
IQR=quantile.iloc[0]-quantile.iloc[1]
```

```python
upper_extreme=quantile.iloc[0]+(1.5*IQR)
upper_extreme
```

```python
lower_extreme=quantile. iloc[1]-(1.5* IQR)
lower_extreme
```

```python
df.shape
```

```python
df = df[df['price'] <= upper_extreme[0]]
df = df[df['price'] >= lower_extreme[0]]
df = df[df['area'] <= upper_extreme[1]]
df = df[df['area'] >= lower_extreme[1]]
df = df[df['bedrooms'] <= upper_extreme[2]]
df = df[df['bedrooms'] >= lower_extreme[2]]
df = df[df['bathrooms'] <= upper_extreme[3]]
df = df[df['bathrooms'] >= lower_extreme[3]]
df = df[df['stories'] <= upper_extreme[4]]
df = df[df['stories'] >= lower_extreme[4]]
df = df[df['parking'] <= upper_extreme[5]]
df = df[df['parking'] >= lower_extreme[5]]
df.shape
```

#### Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.head()
```

```python
lst = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','furnishingstatus']

le = LabelEncoder()
df[lst] = df[lst].apply(le.fit_transform)
```

```python
df.head()
```

#### Split the data into dependent and independent variables

```python
x=df.iloc[:,1:12]
x.head()
```

```python
y=df.iloc[:,0:1]
y.head()
```

#### Scale the independent variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
x1 = StandardScaler().fit_transform(x)
```

####  Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.20, random_state=42)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression().fit(X_train, y_train)
y_pred1 = regr.predict(X_test)

print("R2 Score: ", r2_score(y_pred1,y_test))

print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(y_test,y_pred1)))
```

#### Ridge Regression

```python
from sklearn.linear_model import Ridge
r=Ridge().fit(X_train,y_train)
y_pred2 = r.predict(X_test)

print("R2 Score: ", r2_score(y_pred2,y_test))

print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(y_test,y_pred2)))
```

#### Lasso Regression

```python
from sklearn.linear_model import Lasso
l=Lasso().fit(X_train,y_train)
y_pred3=l.predict(X_test)

print("R2 Score: ", r2_score(y_pred3,y_test))

print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(y_test,y_pred3)))
```
